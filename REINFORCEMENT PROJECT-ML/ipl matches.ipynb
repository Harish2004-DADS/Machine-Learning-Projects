```python
ipl.head()
```

```python
ipl['Season'] = ipl['Season'].str.replace("IPL-", "", regex=False)
ipl.drop('Id',axis=1,inplace=True)
```

```python
ipl.head()
```

- LABEL ENCODER

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
# x=ipl[['Team1','Team2','Toss Winner','Toss Decision','Result','Win By Runs','Win By Wickets','Player Of Match','Venue']]
x=ipl[['City','Team1','Team2','Toss Winner','Toss Decision']]
y=ipl['Winner']
```

```python
encoders = {}
x_encoded = x.copy()

for col in x.columns:
    le = LabelEncoder()
    x_encoded[col] = le.fit_transform(x[col])
    encoders[col] = le
```

```python
le_y = LabelEncoder()
y_encoded = le_y.fit_transform(y)
```

- TRAIN & TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y_encoded, test_size = 0.2, random_state = 42)
```

```python
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# x_train = sc.fit_transform(x_train)
# x_test = sc.transform(x_test)
```

### LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score,precision_score, recall_score, f1_score
```

```python
param_grid = {
    "C": [0.01, 0.1, 1, 10, 100],
    "solver": ["liblinear", "lbfgs"],
    "penalty": ["l1", "l2"]
}
```

```python
logi = GridSearchCV(LogisticRegression(max_iter=1000), param_grid, cv=5, scoring="accuracy", n_jobs=-1)
logi.fit(x_train, y_train)
```

```python
best_lr = logi.best_estimator_
y_train_pred_log = best_lr.predict(x_train)
y_test_pred_log = best_lr.predict(x_test)
```

```python
accuracy_train_log=accuracy_score(y_train,y_train_pred_log)
accuracy_test_log=accuracy_score(y_test,y_test_pred_log)
precision_log=precision_score(y_test, y_test_pred_log,average="weighted")
recall_log=recall_score(y_test, y_test_pred_log,average="weighted")
f1_log=f1_score(y_test, y_test_pred_log,average="weighted")
```

```python
print('Accuracy Score Logistic Train:',accuracy_train_log)
print('Accuracy Score Logistic Test:',accuracy_test_log)
print('Precision Score Logistic:',precision_log)
print('Recall Score Logistic:',recall_log)
print('F1 Score Logistic:',f1_log)
```

```python
confusion_matrix(y_test,y_test_pred_log)
```

### KNN (K- Nearest Neighbors) Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import  cross_val_score, KFold
```

```python
k_range = range(1, 21) 
cv = KFold(n_splits=10, shuffle=True, random_state=42) 
mean_mse = [] 

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')  
    mse = cross_val_score(knn, x_train, y_train, cv=cv, scoring='neg_mean_squared_error')
    mean_mse.append(-mse.mean())  
```

```python
best_k = k_range[np.argmin(mean_mse)]
print(f'Best K: {best_k}')
```

```python
knn = KNeighborsClassifier(n_neighbors=best_k, metric='euclidean')
```

```python
param_grid = {
    "n_neighbors": [3, 5, 7, 9, 11, 15, 20],   # number of neighbors
    "weights": ["uniform", "distance"],        # how neighbors are weighted
    "metric": ["euclidean", "manhattan", "minkowski"]  # distance metrics
}
```

```python
gridknn = GridSearchCV(
    estimator=knn,
    param_grid=param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1,
    verbose=2
)
```

```python
gridknn.fit(x_train, y_train)
```

```python
y_train_pred_knn = gridknn.predict(x_train)
y_test_pred_knn = gridknn.predict(x_test)
```

```python
accuracy_train_knn=accuracy_score(y_train,y_train_pred_knn)
accuracy_test_knn=accuracy_score(y_test,y_test_pred_knn)
precision_knn=precision_score(y_test, y_test_pred_knn,average="weighted")
recall_knn=recall_score(y_test, y_test_pred_knn,average="weighted")
f1_knn=f1_score(y_test, y_test_pred_knn,average="weighted")
```

```python
print('Accuracy Score KNN Train:',accuracy_train_knn)
print('Accuracy Score KNN Test:',accuracy_test_knn)
print('Precision Score KNN:',precision_knn)
print('Recall Score KNN:',recall_knn)
print('F1 Score KNN:',f1_knn)
```
